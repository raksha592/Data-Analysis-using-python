## 1. Plot of percentage of Bachelor's degree conferred by women through the years in USA
## 2. Plot of percentage of women enrolled into various majors and the average IQ by all gender in each major field

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("~/Final/Data/percent-bachelors-degrees-women-usa.csv")
df.head()
```

```python
# Line Plot of Percentage of Women Vs Year in each major field

plt.close('all')
df3=df.set_index('Year')
fig, ax = plt.subplots(1, 1, figsize=(25, 25))
 
ax.plot(df3)
 
ax.set(xlabel='Year',ylabel='Percentage Of Women in each major',title='Percentage of Bachelors degree conferred by women in USA')
 
ax.legend(df3.columns, loc=4)
fig = ax.get_figure()
fig.savefig('PercentageOfWomenVsYear')
ax.figure
```

```python
df1=pd.read_csv("~/Final/Data/iq-vs-salary-by-major.csv")
```

```python
#AverageIQ vs percentage of Women scatter plot
major=[]
for index,row in df1.iterrows():
    major.append(row['Major'])
df2=pd.DataFrame({'Average_IQ': df1.Average_IQ,'Percentage_Female':df1.Pct_Female,'Major':major})
fg = sns.FacetGrid(data=df2, hue='Major', hue_order=major, aspect=1.61)
fg.map(plt.scatter, 'Average_IQ', 'Percentage_Female').add_legend()
fg.savefig("Percentage_femaleVsAverage_IQ")
fg.fig
```
