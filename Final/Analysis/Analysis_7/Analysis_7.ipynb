### Analysis 7
+ Plot 1: Scatter plot of students with highest Verbal SAT
+ Plot 2: Scatter plot of students with highest Quant SAT
+ Plot 3: Scatter plot of students with highest Average SAT

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1=pd.read_csv("~/Final/Data/iq-vs-salary-by-major.csv")
df1.head()
```

```python
major=[]
for index,row in df1.iterrows():
    major.append(row['Major'])
df2=pd.DataFrame({'Verbal SAT score': df1.Verbal_SAT,'Total candidates':df1.Bachelors_Total,'Major':major})
fg = sns.FacetGrid(data=df2, hue='Major', hue_order=major, aspect=1.61)
fg.map(plt.scatter, 'Verbal SAT score', 'Total candidates').add_legend()
fg.savefig("VerbalSAT")
fg.fig
```

```python
df3=pd.DataFrame({'Quantitative SAT score': df1.Quant_SAT,'Total candidates':df1.Bachelors_Total,'Major':major})
fg = sns.FacetGrid(data=df3, hue='Major', hue_order=major, aspect=1.61)
fg.map(plt.scatter, 'Quantitative SAT score', 'Total candidates').add_legend()
fg.savefig("QuantSAT")
fg.fig
```

```python
df4=pd.DataFrame({'Total SAT': df1.Average_SAT,'Total candidates':df1.Bachelors_Total,'Major':major})
fg = sns.FacetGrid(data=df4, hue='Major', hue_order=major, aspect=1.61)
fg.map(plt.scatter, 'Total SAT', 'Total candidates').add_legend()
fg.savefig("AverageSAT")
fg.fig
```
