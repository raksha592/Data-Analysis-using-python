# Analysis 3

## Plot 1: Unemployment rates comparision between grads and non-grads
## Plot 2: Number of candidates employed among Grads and Nongrads
## Plot 3: Number of candidates employed with full-time and halftime among grads
## Plot 4: Number of candidates employed with full-time and halftime among Non grads

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1=pd.read_csv("~/Final/Data/grad-students.csv")
df1.head()
```

```python
#creating a list of all grads and non grads in each major category
major=list(df1.groupby(by=['Major_category'])['Major_category'].unique())
gradTotal=list(df1.groupby(by=['Major_category'])['Grad_total'].sum())
NongradTotal=list(df1.groupby(by=['Major_category'])['Nongrad_total'].sum())
```

```python
gradEmployed=list(df1.groupby(by=['Major_category'])['Grad_employed'].sum())
gradEmployedFullTime=list(df1.groupby(by=['Major_category'])['Grad_full_time_year_round'].sum())
gradUnemployed=list(df1.groupby(by=['Major_category'])['Grad_unemployed'].sum())
for i in range(0,len(gradEmployed)):
    gradEmployedHalfTime[i]=gradEmployed[i]-gradEmployedFullTime[i] #Students employed halftime
    gradNotGraduated[i]=gradTotal[i]-(gradEmployed[i]+gradUnemployed[i]) #students who did not graduate
    gradUnemploymentRate[i]=gradUnemployed[i]/(gradEmployed[i]+gradUnemployed[i]) #Unemployment rates among students who graduated from college
gradEmployedHalfTime=list(gradEmployedHalfTime)
gradNotGraduated=list(gradNotGraduated)
gradUnemploymentRate=list(gradUnemploymentRate)
```

```python
NongradEmployed=df1.groupby(by=['Major_category'])['Nongrad_employed'].sum()
NongradEmployedFullTime=df1.groupby(by=['Major_category'])['Nongrad_full_time_year_round'].sum()
NongradUnemployed=df1.groupby(by=['Major_category'])['Nongrad_unemployed'].sum()
for i in range(0,len(NongradEmployed)):
    NongradEmployedHalfTime[i]=NongradEmployed[i]-NongradEmployedFullTime[i] #Students employed halftime
    NongradUnemploymentRate[i]=NongradUnemployed[i]/(NongradEmployed[i]+NongradUnemployed[i]) #Unemployment rates among students who did not go to college
NongradEmployedHalfTime=list(NongradEmployedHalfTime)
NongradUnemploymentRate=list(NongradUnemploymentRate)
```

```python
#plot of unemployment rates in each major
df2=pd.DataFrame({'Major_category':major, 'Unemployment rates:Grad Students':gradUnemploymentRate, 'Unemployment rates:NonGrad Students':NongradUnemploymentRate})
df2=df2.set_index('Major_category')
df2.head()
```

```python
plt.close('all')
ax = df2[['Unemployment rates:Grad Students','Unemployment rates:NonGrad Students']].plot(kind='bar', title ="Unemployment Rates", legend=True, fontsize=12)
ax.set_xlabel("Major Category", fontsize=12)
ax.set_ylabel("Unemployment rates", fontsize=12)
fig = ax.get_figure()
fig.savefig("UnemploymentRatesGradsVsNongrads")
plt.show()
```

```python
#plot of employed in grad vs non grad
df3=pd.DataFrame({'Major_category':major, 'Grad Employed':gradEmployed, 'Nongrad Employed':NongradEmployed})
df3=df3.set_index('Major_category')
df3.head()
```

```python

plt.close('all')
ax = df3[['Grad Employed','Nongrad Employed']].plot(kind='bar', title ="Grad Employed Vs Nongrad Employed", legend=True, fontsize=12)
ax.set_xlabel("Major Category", fontsize=12)
ax.set_ylabel("Total Employed", fontsize=12)
fig = ax.get_figure()
fig.savefig("GradEmployedVsNongradEmployed")
plt.show()
```

```python
#plot of fulltime,halftime in grads
df4=pd.DataFrame({'Major_category':major, 'Grad Employed full-time':gradEmployedFullTime, 'Grad Employed half-time':gradEmployedHalfTime})
df4=df4.set_index('Major_category')
df4.head()
```

```python
plt.close('all')
ax = df4[['Grad Employed full-time','Grad Employed half-time']].plot(kind='bar', title ="Grad Employed Full-time Vs Grad Employed Half-time", legend=True, fontsize=12)
ax.set_xlabel("Major Category", fontsize=12)
ax.set_ylabel("Total Employed", fontsize=12)
fig = ax.get_figure()
fig.savefig("GradEmployedFullTimeVsGradEmployedhalfTime")
plt.show()
```

```python
#plot of full time, halftime in nongrads
df5=pd.DataFrame({'Major_category':major, 'NonGrad Employed full-time':NongradEmployedFullTime, 'NonGrad Employed half-time':NongradEmployedHalfTime})
df5=df5.set_index('Major_category')
df5.head()
```

```python
plt.close('all')
ax = df5[['NonGrad Employed full-time','NonGrad Employed half-time']].plot(kind='bar', title ="Non Grad Employed Full-time Vs Non Grad Employed Half-time", legend=True, fontsize=12)
ax.set_xlabel("Major Category", fontsize=12)
ax.set_ylabel("Total Employed", fontsize=12)
fig = ax.get_figure()
fig.savefig("NonGradEmployedFullTimeVsNonGradEmployedhalfTime")
plt.show()
```
