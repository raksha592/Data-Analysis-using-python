### Analysis 6
+ Comparison of Total graduated and Total employed in each major

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df1=pd.read_csv('~/Final/Data/grad-students.csv')
df1.head()
```

```python
major=list(df1.groupby(by=['Major_category'])['Major_category'].unique())
```

```python
grad_total=list(df1.groupby(by=['Major_category'])['Grad_total'].sum())
```

```python
df2=pd.DataFrame({'Major_Graduated':major,'Total Graduated':grad_total})
```

```python
df2=df2.sort_values(by=['Total Graduated'],ascending=[False])
df2
```

```python
gradEmployed=list(df1.groupby(by=['Major_category'])['Grad_employed'].sum())
```

```python
df3=pd.DataFrame({'Major_Employed':major,'Total Employed':gradEmployed})
df3=df3.sort_values(by=['Total Employed'],ascending=[False])
df3
```

```python
df4=pd.merge(df2, df3, right_index=True, left_index=True)
df4.to_csv('TotalGraduatedVsTotalEmployed.csv', encoding='utf-8')
```
