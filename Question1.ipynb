```python
import nltk
from nltk.corpus import gutenberg
import collections
import re
from collections import Counter
gutenberg.fileids()
```

```python
words=gutenberg.words('austen-persuasion.txt')
N=len(words) #number of words in austen-persuasion.txt
UN=len(set(words)) #number of unique words
print("The number of words in austen-persuasion.txt is ",N)
print("The number of words in austen-persuasion.txt is ",UN)
```

```python
lower_word= [word.lower() for word in words]
wordcount={}
for wordNew in lower_word:
    if wordNew not in wordcount:
        wordcount[wordNew]=1
    else:
        wordcount[wordNew]+=1
print("The frequency of each word is:")
for k,v in wordcount.items():
    print(k,v)
```
